```python
# IMPORT LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

%matplotlib inline

# 1) LOAD DATA
df = pd.read_csv("movie_metadata.csv")

print("First 5 rows:")
display(df.head())

print("\nMissing values:")
display(df.isnull().sum())

print("\nSummary statistics for numerical variables:")
display(df.describe())

# 2) BASIC EDA PLOTS
plt.figure(figsize=(8,5))
sns.histplot(df['imdb_score'], bins=20, kde=True)
plt.title("IMDb Score Distribution")
plt.xlabel("IMDb Score")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df['budget'].dropna(), bins=30)
plt.title("Budget Distribution")
plt.xlabel("Budget")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='budget', y='gross')
plt.title("Budget vs Gross Revenue")
plt.xlabel("Budget")
plt.ylabel("Gross Revenue")
plt.show()

plt.figure(figsize=(10,5))
sns.countplot(data=df.dropna(subset=['title_year']), x='title_year')
plt.xticks(rotation=90)
plt.title("Number of Films per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()

# 3) GENRE ANALYSIS
df['main_genre'] = df['genres'].str.split('|').str[0]

plt.figure(figsize=(10,5))
df['main_genre'].value_counts().plot(kind='bar')
plt.title("Main Genre Distribution")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()

print("\nAverage IMDb Score by Main Genre:")
display(df.groupby('main_genre')['imdb_score'].mean().sort_values(ascending=False))

# 4) HYPOTHESIS TESTS

print("\n----------- H1: Drama vs Action IMDb Scores -----------")
drama = df[df['main_genre']=="Drama"]['imdb_score'].dropna()
action = df[df['main_genre']=="Action"]['imdb_score'].dropna()

t_stat1, p_val1 = ttest_ind(drama, action, equal_var=False)
print(f"Drama mean: {drama.mean():.3f}, Action mean: {action.mean():.3f}")
print(f"t-statistic: {t_stat1:.3f}, p-value: {p_val1:.5f}")

if p_val1 < 0.05:
    print("→ Reject H0. IMDb scores are significantly different between Drama and Action.")
else:
    print("→ Fail to reject H0. No significant difference between Drama and Action.")

print("\n----------- H2: Pre-2000 vs Post-2000 IMDb Scores -----------")
before_2000 = df[df['title_year'] < 2000]['imdb_score'].dropna()
after_2000  = df[df['title_year'] >= 2000]['imdb_score'].dropna()

t_stat2, p_val2 = ttest_ind(before_2000, after_2000, equal_var=False)
print(f"Pre-2000 mean: {before_2000.mean():.3f}, Post-2000 mean: {after_2000.mean():.3f}")
print(f"t-statistic: {t_stat2:.3f}, p-value: {p_val2:.5f}")

if p_val2 < 0.05:
    print("→ Reject H0. IMDb scores differ significantly between pre-2000 and post-2000 films.")
else:
    print("→ Fail to reject H0. No significant difference.")

print("\n----------- H3: High-Budget vs Low-Budget IMDb Scores -----------")
median_budget = df['budget'].median()

high_budget = df[df['budget'] >= median_budget]['imdb_score'].dropna()
low_budget  = df[df['budget'] < median_budget]['imdb_score'].dropna()

t_stat3, p_val3 = ttest_ind(high_budget, low_budget, equal_var=False)
print(f"High budget mean: {high_budget.mean():.3f}, Low budget mean: {low_budget.mean():.3f}")
print(f"t-statistic: {t_stat3:.3f}, p-value: {p_val3:.5f}")

if p_val3 < 0.05:
    print("→ Reject H0. IMDb scores differ significantly between high and low budget films.")
else:
    print("→ Fail to reject H0. No significant difference.")

# 5) CORRELATION HEATMAP
plt.figure(figsize=(8,6))
sns.heatmap(df[['imdb_score','budget','gross','title_year']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
